## Louvain 

```python
def louvain_performance():
    for net in networks_und:
        print("Louvain for undirected " + net.name.split("/")[2])
        start_time = time.time()
        ScLouvain = algorithms.louvain(net)
        t = time.time() - start_time
        print("Execution time: %.6s seconds" % (t))
        
        coms = list(ScLouvain.to_node_community_map().values())[-1][0]+1
        print("Number of communities:", coms)
        ACS = evaluation.size(net, ScLouvain)
        print("Average Community Size:", ACS[2])
        AID = evaluation.average_internal_degree(net, ScLouvain)
        print("Average Internal Degree:", AID[2])

        NG_Mod = evaluation.newman_girvan_modularity(net, ScLouvain)
        print("Newman-Girvan Modularity:", NG_Mod[2])
        ER_Mod = evaluation.erdos_renyi_modularity(net, ScLouvain)
        print("Erdos-Renyi Modularity:", ER_Mod[2])
        Z_Mod = evaluation.z_modularity(net, ScLouvain)
        print("Z-Modularity:", Z_Mod[2])
        Link_Mod = evaluation.link_modularity(net, ScLouvain)
        print("Link Modularity:", Link_Mod[2])
        significance = evaluation.significance(net, ScLouvain)
        print("Significance:", significance[2])
        print()
louvain_performance()
```
